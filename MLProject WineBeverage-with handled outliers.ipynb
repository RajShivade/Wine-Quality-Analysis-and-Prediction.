### `Dataset description` :

The dataset contains 6498 rows and 14 columns:

- `fixed acidity` – It is a acids like tartaric acid; contributes to taste & structure.

- `volatile acidity` – It is a high values make wine taste vinegary (acetic acid).

- `citric acid` – It is a adds freshness & citrus flavor.

- `chlorides` – It is a salt content; affects mouthfeel.

- `free sulfur dioxide` – It is a prevents spoilage.

- `total sulfur dioxide` – It is a sum of free + bound SO₂; excess impacts aroma.

- `density` – It is a higher density often means higher sugar.

- `pH`– It is a acidity level (low pH = more acidic).

- `sulphates` – It is a enhance preservation & flavor.

- `alcohol` – It is a % of alcohol by volume; strongly linked to quality.

- `quality` – It is a sensory rating of wine (0–10) → Target variable.

- `good (1/0)` – It is a binary label (1 = good quality, 0 = not good).

- `color` – It is a wine type (red/white).

### Handling Outliers :

`Finding Unique of all column`


```python
df.skew(numeric_only =True)
```

```python
def get_IQRValues(data,column):
    q1,q2,q3 =data[column].quantile([0.25,0.5,0.75])
    iqr = q3 -q1
    upper_limit = q3 + 1.5 *iqr
    lower_limit = q1 - 1.5 * iqr
    return upper_limit ,lower_limit
```

```python
import pandas as pd

# Function to remove outliers using IQR
def remove_outliers_iqr(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Filter the dataframe
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

# Columns with high skewness (likely outliers)
skewed_cols = ['fixed acidity', 'volatile acidity', 'residual sugar', 
               'chlorides', 'free sulfur dioxide', 'sulphates']

# Remove outliers
df_clean = remove_outliers_iqr(df, skewed_cols)

print("Original shape:", df.shape)
print("After removing outliers:", df_clean.shape)
```

- Grapes & Soil

If grapes are grown in soil with higher salinity (salt content), the grapes absorb more chlorides.

This is common in vineyards near the sea or saline water sources.

- Winemaking Process

Red wine is fermented with the skins (unlike many white wines).

The contact with skins and seeds can release more minerals, including chlorides, into the wine.

- Aging & Storage

Sometimes chloride levels can increase depending on the type of barrels or storage conditions.

- Preservation of Taste

A slightly higher chloride content can contribute to the body and taste balance of red wine.

```python
grouped = df_clean.groupby('quality_category')[['pH', 'alcohol']].mean().reset_index()

# Set position of bars on X axis
brands = grouped['quality_category']
x = np.arange(len(brands))
width = 0.35

# Plotting
fig, ax1 = plt.subplots(figsize=(12, 6))

# Primary axis for Price
bar1 = ax1.bar(x - width/2, grouped['pH'], width, label='pH', color='red')
ax1.set_ylabel('pH', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Secondary axis for ABV%
ax2 = ax1.twinx()
bar2 = ax2.bar(x + width/2, grouped['alcohol'], width, label='alcohol', color='gold')
ax2.set_ylabel('alcohol', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# X-axis settings
ax1.set_xlabel('quality_category')
ax1.set_xticks(x)
ax1.set_xticklabels(brands, rotation=45)

# Title
plt.title('Comparison of pH and alcohol by quality_category')

# Legends
fig.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)

plt.tight_layout()
plt.show()
```
